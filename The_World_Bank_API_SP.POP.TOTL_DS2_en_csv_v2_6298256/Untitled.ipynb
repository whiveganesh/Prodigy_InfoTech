### Identify countries or regions that experienced the highest or lowest population growth during specific periods.


idxmin() returns the index where the minimum value occurs.


idxmax() returns the index where the maximum value occurs.

```python
periods_of_interest = [(1960, 1970), (1971, 1980), (1981, 1990), (1991, 2000), (2001, 2010), (2011, 2022)]
results = pd.DataFrame(columns=['Period', 'Country Name', 'Population Growth Rate'])
# Create an empty list to store DataFrames
result_dfs = []

for period in periods_of_interest:
    start_year, end_year = period
    period_name = f'{start_year}-{end_year}'

    # Calculate growth rate for the period
    df[f'Growth Rate_{period_name}'] = ((df[str(end_year)] - df[str(start_year)]) / df[str(start_year)]) * 100

    # Identify country with the highest growth rate
    highest_growth_country = df.loc[df[f'Growth Rate_{period_name}'].idxmax(), 'Country Name']
    highest_growth_rate = df.loc[df[f'Growth Rate_{period_name}'].idxmax()][f'Growth Rate_{period_name}']

    # Identify country with the lowest growth rate
    lowest_growth_country = df.loc[df[f'Growth Rate_{period_name}'].idxmin(), 'Country Name']
    lowest_growth_rate = df.loc[df[f'Growth Rate_{period_name}'].idxmin()][f'Growth Rate_{period_name}']


    # # Print information for the current period
    # print(f"Period: {period_name}")
    # print(f"Country with the Highest Growth: {highest_growth_country}, Growth Rate: {highest_growth_rate}")
    # print(f"Country with the Lowest Growth: {lowest_growth_country}, Growth Rate: {lowest_growth_rate}")
    # print()

    # Create a DataFrame for the current period
    period_df = pd.DataFrame({
        'Period': period_name,
        'Country Name (Highest Growth)': highest_growth_country,
        'Population Growth Rate (Highest)': highest_growth_rate,
        'Country Name (Lowest Growth)': lowest_growth_country,
        'Population Growth Rate (Lowest)': lowest_growth_rate
    }, index=[0])

    # Append the DataFrame to the list
    result_dfs.append(period_df)
```

```python
# Concatenate all DataFrames in the list
results = pd.concat(result_dfs, ignore_index=True)
results
```

### Rank countries based on their population size for specific years.

```python
# Select relevant data for the specified range of years (1960 to 1970)
selected_data = df[['Country Name', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970']]

# Calculate the average population for the specified years
selected_data['Average Population(1960-1970'] = selected_data[['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970']].mean(axis=1)

# Rank countries based on the average population
selected_data['Rank'] = selected_data['Average Population(1960-1970'].rank(ascending=False)

# Sort the DataFrame by rank
selected_data = selected_data.sort_values(by='Rank')

# Display the ranked data
print(selected_data[['Country Name', 'Average Population(1960-1970', 'Rank']])
```

```python
# Visualize changes in rankings over time for the top 10 countries
plt.figure(figsize=(10, 6))

# Select the top 10 countries
top_10_countries = selected_data.head(22)

for country in top_10_countries['Country Name']:
    country_data = selected_data[selected_data['Country Name'] == country]
    plt.plot(country_data['Average Population(1960-1970'], country_data['Rank'], label=country, marker='o')

plt.title('Average Population Rankings (1960 to 1970) - Top 10 Countries')
plt.xlabel('Average Population(1960-1970')
plt.ylabel('Rank')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Select relevant data for the specified range of years (2000 to 2022)
selected_data_2000_2022 = df[['Country Name'] + list(df.loc[:, '2000':'2022'].columns)]

# Calculate the average population for the specified years
selected_data_2000_2022['Average Population(2000-2022)'] = selected_data_2000_2022.loc[:, '2000':'2022'].mean(axis=1)

# Rank countries based on the average population
selected_data_2000_2022['Rank'] = selected_data_2000_2022['Average Population(2000-2022)'].rank(ascending=False)

# Sort the DataFrame by rank
selected_data_2000_2022 = selected_data_2000_2022.sort_values(by='Rank')

# Display the ranked data
print(selected_data_2000_2022[['Country Name', 'Average Population(2000-2022)', 'Rank']])
```
